Hyphens are found to:  
1. Split the name to metadata and activity  
'A10 (34/2013, 25 Aug 13) - Implementation of Specific Plans'

'Transport TBM components from site to the ferry pier at night - 623N2' 
'855 - Kwai Chung Vehicular Access Adit Rooms'

'Twin track tunnel drill & blast Ch134979 - Ch135279 (300m)'

'Tunnel invert & Walkway(UT-Northbound) (Ch136+475~Ch136+656) (Bay F02 - F01)'
