```python
import os
import re
import nltk

def isfloat(value):
    '''
    Check if the input value type is float
    '''
    try:
        float(value)
        return float(value)
    except ValueError:
        return False

def isint(value):
    '''
    Check if the input value type is integer
    '''
    try:
        int(value)
        return int(value)
    except ValueError:
        return False
tokens_types = {} 
```

```python
results_dir = 'C:\\Users\\RonyArmon\\Projects_Code\\Cluster_Activities\\results'
unique_names = open(os.path.join(results_dir,'names.txt')).read().split('\n')
print('{n} unique names tokenized'.format(n=len(unique_names)))
tokens = []
for name in unique_names:
    name_tokens = nltk.word_tokenize(name)
    tokens += name_tokens
tokens = list(set(tokens))
print('{} tokens'.format(len(tokens)))

# Split by punctuation: backslah and hyphen 
split_backslash, split_hyphen = [], []
for t in tokens:
    ts = t.split('/')
    split_backslash += ts
for t in split_backslash:
    ts = t.split('-')
    split_hyphen += ts
tokens = [t for t in split_hyphen if len(t)>1]
print('{} tokens following punctuation split'.format(len(tokens)))

# Tokens types
numeric_tokens = [t for t in split_hyphen if type(isfloat(t))==float]
print('{} numeric_tokens'.format(len(numeric_tokens)))
```

```python
metric_tokens = [t for t in split_hyphen if re.findall('^[+|-]?\d{1,}\.?\d{1,}[a-z]{1,}$', t)]
print('{} metric_tokens'.format(len(metric_tokens)))
print(metric_tokens)
```

```python
with_meter = [t for t in split_hyphen if re.findall('^[+|-]?\d{1,}\.?\d{1,}m$', t)] 
print('{} with meter'.format(len(with_meter)))
print(with_meter)
```

```python
with_v = [t for t in split_hyphen if re.findall('^[+|-]?\d{1,}\.?\d{1,}v$', t)] 
print('{} with v'.format(len(with_v)))
print(with_v)
```

```python
with_mpd = [t for t in split_hyphen if re.findall('^[+|-]?\d{1,}\.?\d{1,}mpd$', t)] 
print('{} with v'.format(len(with_mpd)))
print(with_mpd)
```

```python
metric_names = []
for name in unique_names:
    if any(t in name for t in with_mpd):
        print(name)
```

```python
none_numeric = [t for t in split_hyphen if t not in numeric_tokens] 
print('{} none_numeric tokens'.format(len(none_numeric)))
print(none_numeric)
```

```python
dot_tokens = [t for t in none_numeric if '.' in t]
print('{} none-numeric tokens contain dots'.format(len(dot_tokens)))
print(dot_tokens)
```
