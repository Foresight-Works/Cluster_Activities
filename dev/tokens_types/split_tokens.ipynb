Tokenization product, further split on backslah, dot or hyphen

```python
import re
```

```python
tokens = open('tokens.txt').read().split('\n')
print('{} tokens'.format(len(tokens)))
```

```python
split1, split2, split3, split4 = [], [], [], []
for token in tokens: split1 += token.split('/')
for token in split1: split2 += token.split('-')
for token in split2: split3 += token.split('.')
for token in split3: split4 += token.split(',')
splitted = [t for t in split4 if len(t)>1]
print(splitted)
```

```python
no_numbers = [re.sub('\d','',t) for t in splitted]
no_numbers = [t for t in no_numbers if len(t)>1]
print(no_numbers)
```
