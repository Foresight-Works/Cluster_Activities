Tokenization product, further split on backslah, dot or hyphen

```python
import re
```

```python
tokens = open('tokens.txt').read().split('\n')
print('{} tokens'.format(len(tokens)))
```

```python
split1, split2, split3, split4 = [], [], [], []
for token in tokens: split1 += token.split('/')
for token in split1: split2 += token.split('-')
for token in split2: split3 += token.split('.')
for token in split3: split4 += token.split(',')
splitted = [t for t in split4 if len(t)>1]
print(splitted)
```

```python
letters_only = [t for t in tokens if re.sub('[a-z]', '', t)=='']
print('{} tokens that contain only letters'.format(len(letters_only)))
print(letters_only)
```

```python
mixed_chars = [t for t in tokens if t not in letters_only]
print('{} mixed character tokens'.format(len(mixed_chars)))
print(mixed_chars)
```

```python
with_backslash = [t for t in mixed_chars if '/' in t]
print('{} mixed character with backslash'.format(len(with_backslash)))
print(with_backslash)
```
