## Parenthesized expressions 

```python
s = 'S624 Excavate 92m (SZ) (c.rock/total 1864m)'
pattern = '\w+|\$[\d\.]+'
pattern = '\w+'
pattern = '\(.+?\)|\w\d{1,}|\d{1,}\w'
pattern = '\(.+?\)|\d{1,}\w'
pattern = '\(.+?\)|\w+\d{1,}\w+|\w+'
pattern = '\(.+?\)|\w+\d{1,}\.*\d*\w*|\w+'
tokenizer = RegexpTokenizer(pattern)
print(s)
print(tokenizer.tokenize(s))
```
